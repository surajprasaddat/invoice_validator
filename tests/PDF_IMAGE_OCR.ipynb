````python
def ocr_pdf(file_path):
    """
    Perform OCR on a PDF and return extracted text
    Works for both image-based and text-based PDFs
    """
    text_output = ""
    pdf = fitz.open(file_path)

    for page_num, page in enumerate(pdf, start=1):
        # Try extracting text normally
        text = page.get_text()

        if text.strip():  # If PDF has extractable text
            text_output += f"\n--- Page {page_num} ---\n{text}"
        else:  # If PDF is scanned image, use OCR
            pix = page.get_pixmap(dpi=300)
            img = Image.open(io.BytesIO(pix.tobytes("png")))
            ocr_text = pytesseract.image_to_string(img)
            text_output += f"\n--- Page {page_num} (OCR) ---\n{ocr_text}"

    return text_output
````
